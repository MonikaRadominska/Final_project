### Importing data from IUCN's Red List

```python
df_redlist = pd.read_csv('data/assessments.csv')
#df_redlist is the red list assessments of IUCNs species

df_taxom = pd.read_csv('data/taxonomy.csv')
#df_taxom is the taxonomy df where we look at specific types, classes, family names 
```

```python
#dropping columns in redlist df
df_redlist = df_redlist.drop(['assessmentId', 'internalTaxonId','redlistCriteria','assessmentDate','criteriaVersion',
                              'language', 'rationale', 'habitat', 'threats', 'population',
                              'populationTrend', 'range', 'useTrade','conservationActions',
                              'yearLastSeen', 'possiblyExtinct','possiblyExtinctInTheWild'], axis = 1)

df_redlist
```

```python
def rename_columns(x):
    x.rename(columns={'scientificName':'scientific_name','redlistCategory':'red_list_category', "yearPublished":"year_published", 
                          'realm':'region', "scopes":"habitat"}, inplace=True )
    return x
```

```python
df_redlist = rename_columns(df_redlist)
```

```python
df_taxom = df_taxom.drop(['internalTaxonId', 'kingdomName', 'phylumName',
                          'orderName', 'familyName', 'genusName', 'speciesName',
                          'infraType', 'infraName', 'infraAuthority', 'subpopulationName',
                          'authority', 'taxonomicNotes'], axis = 1)
```

```python
def rename_columns(x):
    x.rename(columns={'scientificName':'scientific_name','className':'class_name'}, inplace=True )
    return x
```

```python
df_taxom = rename_columns(df_taxom)
```

```python
df_taxom.head()
```

```python
iucn_df = pd.merge(df_redlist, df_taxom, on='scientific_name')
```

```python
iucn_df
```
