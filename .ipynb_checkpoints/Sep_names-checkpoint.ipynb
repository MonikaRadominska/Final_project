```python
list_names = names['NAMES'].values.tolist()
list_names
```

```python
del list_names[2]
list_names
```

```python
for i in range(len(names)-1):
    if list_names [i].startswith('N'):
        if list_names [i+1].startswith('C'):
            print("no changes")
        else: list_names.insert(i+1, 'unknown')
    if list_names [i].startswith('C'):
        if list_names [i+1].startswith('N'):
            print("no changes")
        else: list_names.insert(i+1, 'unknown')
```

```python
#list slicing

scientific = list_names[0:-1:2]
common = list_names[1:-1:2]
```

```python
df_ok = pd.DataFrame(list(zip(scientific, common)), columns = ['scientific', 'common'])
```

```python
names = names.dropna()
```

```python
names = names.drop("scientific", axis=1)
names
```

```python
names_sep = pd.DataFrame()
names_sep["scientific"] = names[names.apply(lambda x: x.str.startswith("N"))]
names_sep["common"] = names[names.apply(lambda x: x.str.startswith("C"))]
```

```python
names_sep = names_sep.dropna(how="all").fillna("unknown")
```

```python
len(names_sep[names_sep["common"] == "unknown"]["common"])
```
